```python
from PIL import Image
import numpy as np
import gist
from cifar import CIFAR10
import pickle
```

```python
dataset = CIFAR10(dataset_root='./cifar10')
train=dataset.samples['train']
pickle_out = open("gist.cifar10","wb")
```

```python
gist_images=[]
for i in train:
    descriptor = gist.extract(i.image)   
    gist_images.append(descriptor)

pickle.dump(gist_images, pickle_out)
pickle_out.close()
#print(gist_images)
```

```python
pickle_in = open("gist.cifar10","rb")
gistImages = pickle.load(pickle_in)
print(gistImages)
```
